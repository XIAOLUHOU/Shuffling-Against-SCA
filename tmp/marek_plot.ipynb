```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
cur_path = os.path.dirname(os.getcwd())
new_path = os.path.relpath('..\\C\\results.txt', cur_path)
f = open(new_path, "r")
results = f.read().split("\n")[1:-1]
#results = [float(i) for i in results]
```

```python
results2 = list()
for x in results:
    results2.append(x.split(" "))
results = np.array(results2)
results = [float(i) for i in results[:,1]]
```

```python
nums = [j for j in range(50,1001,50)]
plt.plot(nums, results, marker='o', linestyle='-', color='r')
plt.title("Overhead: Fisher Yates Masked over Fisher Yates")
plt.xlabel("Number of neurons")
plt.ylabel("Overhead")
plt.show()
```
